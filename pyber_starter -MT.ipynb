```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
```

```python
# Read the City and Ride Data

city = pd.read_csv(city_data_to_load)
city.head()
```

```python
ride = pd.read_csv(ride_data_to_load)
ride.head()
```

```python
#combined =  ride.join(city, lsuffix='A', rsuffix='B')
#combined =  ride.join(city, how="left", lsuffix='A')
#combined.head()
```

```python
#new_data_set= combined[['cityA','date','fare','ride_id','driver_count','type']]
#new_data_set.head()
```

```python
# Combine the data into a single dataset

combined_ride = pd.merge(ride, city, on="city")

# Display the data table for preview
combined_ride.head()
```

```python
# Obtain the x and y for each of the three city types

urban = combined_ride.loc[combined_ride["type"]=="Urban",:]
rural = combined_ride.loc[combined_ride["type"]=="Rural",:]
suburban = combined_ride.loc[combined_ride["type"]=="Suburban",:]
```

```python
# group the data

urban_group=urban.groupby(["city"])
rural_group=rural.groupby(["city"])
suburban_group=suburban.groupby(["city"])
```

```python
#x

total_urban_rides = urban_group["ride_id"].count()
total_rural_rides = rural_group["ride_id"].count()
total_suburban_rides = suburban_group["ride_id"].count()
```

```python
#y

avg_fare_urban=urban_group["fare"].mean()
avg_fare_rural=rural_group["fare"].mean()
avg_fare_suburban=suburban_group["fare"].mean()
```

```python
#bubble size

driver_count_urban=urban.drop_duplicates("city")
driver_count_rural=rural.drop_duplicates("city")
driver_count_suburban=suburban.drop_duplicates("city")
```

```python
# Build the scatter plots for each city types
plt.scatter(total_urban_rides, avg_fare_urban, s=5*driver_count_urban["driver_count"], alpha=0.50,c="lightcoral",label="Urban",edgecolor="black")
plt.scatter(total_rural_rides, avg_fare_rural, s=5*driver_count_rural["driver_count"], alpha=0.50,c="gold",label="Rural",edgecolor="black")
plt.scatter(total_suburban_rides, avg_fare_suburban, s=5*driver_count_suburban["driver_count"], alpha=0.50,c="lightskyblue",label="Suburban",edgecolor="black")
plt.title("Pyber Ride Sharing Data -2016")
plt.xlabel("Total no. of Rides / Per City")
plt.ylabel("Average Fare ($)")
plt.text(x=42,y=30,s="")
plt.text(x=42,y=30,s=" NOTE ")

plt.grid()
```
